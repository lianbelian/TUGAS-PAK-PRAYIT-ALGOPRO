PRAKTIKUM 1: Modul math

```python
import math
angka = 16
hasil1 = math.sqrt(angka)
hasil2 = math.pow(2, 5)
print("akar kuadrat dari ", angka, "adalah", hasil1)
print("2 pangkat 5 adalah ", hasil2)
```

PRAKTIKUM 2: from ... import ...

```python
from math import sqrt
angka = 25
hasil = sqrt(angka)
print("akar kuadrat dari ", angka, "adalah", hasil)
```

PRAKTIKUM 3: Alias Modul

```python
import numpy as np
data = np.array([1, 2, 3])
print("Isi array =", data)
```

PRAKTIKUM 4: Eksplorasi Modul Standar

```python
import math
import random
import sys
hasil1 = math.sqrt(81)
hasil2 = random.randint(1, 10)
versi = sys.version
print("akar kuadrat dari 81 adalah", hasil1)
print("bilangan bulat acak antara 1 sampai 10 adalah", hasil2)
print("versi python saat ini adalah", versi)
```
