PRAKTIKUM 4: *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Jumlahkan: ", jumlahkan(1,2,3,4,5))
biodata(Nama="Budi", Umur=19, Prodi="TRK")
```

PRAKTIKUM 7: Decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("sebelum fungsi dijalankan.....")
    fungsi()
    print("setelah fungsi dijalankan....")
  return wrapper

@decorator_sapa
def sapa():
  print("halo, mahasiswa POLINES")
sapa()
```

PRAKTIKUM 7: Studi Kasus Rata-rata Nilai

```python
def hitung_rata(data):
  return sum(data) / len(data)

nilai = [80, 90, 85]
rata = hitung_rata(nilai)

print("rata-rata nilai = ", rata)
```

- Studi Kasus Nyata â€“ Program Nilai

```python
def hitung_rata(data):
  if not data:
    return 0
  return sum(data) / len(data)

nilai_list = []

print("Menghitung nilai akhir mahasiswa")
print("1. Isi nilai")
print("2. Selesai")
print(".............................")

pilihan=int(input("pilih 1 atau 2: "))

while pilihan != 2:
  nilai = float(input("Masukkan nilai: "))
  nilai_list.append(nilai)

  pilihan = int(input("pilih 1 atau 2: "))

rata = hitung_rata(nilai_list)
print(f"Rata-rata nilai: {rata}")

if rata >= 60:
    print("lulus")
else:
    print("Tidak lulus")
```
