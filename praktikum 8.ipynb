PRAKTIKUM 4: *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Jumlahkan: ", jumlahkan(1,2,3,4,5))
biodata(Nama="Budi", Umur=19, Prodi="TRK")
```
