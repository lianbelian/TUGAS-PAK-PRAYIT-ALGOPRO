- Modul buatan sendiri

```python
%%writefile konversi.py
def c_to_f(angka):
  f = angka * 1.8 + 32
  print(f"Konversi dari {angka} Celcius ke Fahrenheit adalah: °{f}")

def f_to_c(angka):
  c = (angka - 32 ) / 1.8
  print(f"Konversi dari {angka} Fahrenheit ke Celcius adalah: °{c}")
```

```python
import konversi

print("1. Celcius ke Fahrenheit")
print("2. Fahrenheit ke Celcius")
print()
pilih = int(input("Masukkan angka: "))


if pilih == 1:
  angka =int(input("Masukkan angka yang ingin anda konversi menjadi Fahrenheit: "))
  konversi.c_to_f(angka)
else:
  angka =int(input("Masukkan angka yang ingin anda konversi menjadi Celcius: "))
  konversi.f_to_c(angka)
```

- Studi kasus mini

```python
%%writefile matematika.py
from random import randint

bener_count = 0
salah_count = 0

def penjumlahan(angka):
  global bener_count
  global salah_count
  angka1 = randint(1,101)
  angka2 = randint(1,11)
  tambah = angka1 + angka2
  angka = int(input(str(angka1) + " + " + str(angka2) + " Adalah: "))
  if angka == tambah:
    print("Benar")
    bener_count += 1
  else:
    print("Salah")
    salah_count += 1
    return False


print("Quiz")
print("1. Mau")
print("2. Selesai")
print("...............")

pilih = int(input("Pilih 1 atau 2: "))

while pilih != 2:
  penjumlahan(pilih)
  print("Lanjut ?")
  print("1. Ya")
  print("2. Tidak")
  print("...............")
  pilih = int(input("Pilih 1 atau 2: "))

if pilih == 2:
    print(f"\nTotal benar: {bener_count}")
    print(f"Total salah: {salah_count}")
```
