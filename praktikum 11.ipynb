Praktikum 01 – Sequential Search pada Data Angka



```python
def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan di indeks {i}"
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("data: ", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

Praktikum 02 – Binary Search pada Data Angka

```python
def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left+right) // 2
    if data[mid] == target:
      return f"data {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("data: ", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

Praktikum 03 – Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
  for i in range(len(data)):
    if data[i].lower() == target.lower():
        return f"nama {target} ditemukan di indeks {i}"
  return f"nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```


Praktikum 04 – Studi Kasus Nilai Mahasiswa dengan Binary Search

```python
def binary_search_nilai(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left+right) // 2
    if data[mid] == target:
      return f"nilai {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"nilai {target} tidak ditemukan"

nilai = [60, 70, 75, 80, 85, 90]
print("Data nilai:", nilai)
print(binary_search_nilai(nilai, 85))
print(binary_search_nilai(nilai, 65))
```

Praktikum 05 – Perbandingan Waktu Sequential vs Binary Search

```python
import time

def sequential_search(data, target):
  for i in range(len(data)):
      if data[i] == target:
         return i
  return -1

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left+right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

data = list(range(1, 100000))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("sequential search waktu : ", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("binary search waktu : ", end - start, "detik")
```

Praktikum 06 – Kombinasi Sorting dan Searching

```python
import random

def insertion_sort(data):
    for i in range(1, len(data)):
      key = data[i]
      j = i - 1
      while j >= 0 and data [j] > key:
        data[j + 1] = data[j]
        j -= 1
      data[j + 1] = key
    return data

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
      mid = (left+right) // 2
      if data[mid] == target:
        return f"data {target} ditemukan di indeks {mid}"
      elif data[mid] < target:
        left = mid + 1
      else:
        right = mid - 1
    return f"data {target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("data acak sebelum diurutkan: ", data)
urut = insertion_sort(data)
print("data acak setelah diurutkan: ", urut)
target = urut[7]
print(binary_search(urut, target))
```

**Penugasan 4 - Studi Kasus Barang Toko**

```python
def bubble_sort(data, key_index=1):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][key_index] > data[j + 1][key_index]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

def binary_search(data, target, key_index=1):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid][key_index] == target:
      return f"harga {target} ditemukan di indexs {mid}"
    elif data[mid][key_index] < target:
      left = mid +1
    else:
      right = mid - 1
  return f"harga {target} tidak ditemukan"

data = [
    ("bolpoin", 5000),
    ("origami", 2000),
    ("penghapus", 3000),
    ("busur", 6000),
    ("pensil", 1500)
]
print("Data sebelum dirapihkan", data)
print("Data setelah dirapihkan", bubble_sort(data))
target = int(input("Masukkan harga barang: "))
print(binary_search(data, target))
```

**Penugasan 5 - Analisis Kompleksitas**

```python
import time, random

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1


ukuran_data = [100, 1000, 10000]

print(f"{'Ukuran Data':<15} | {'Sequential (detik)':<20} | {'Binary (detik)':<20}")
print("-" * 55)

for i in ukuran_data:
    data = list(range(i))
    target = i - 1

    # Sequential Search
    mulai = time.perf_counter()
    sequential_search(data, target)
    selesai = time.perf_counter()
    waktu = selesai- mulai

    # Binary Search
    mulai = time.perf_counter()
    binary_search(data, target)
    selesai = time.perf_counter()
    waktu = selesai - mulai

    print(f"{i:<15} | {waktu:<20.8f} | {waktu:<20.8f}")
```

**Penugasan 6 - Proyek Mini**

```python
 def binary_search_nilai(daftar, target):
    right, left = 0, len(daftar) - 1
    while right <= left:
        mid = (right + left) // 2
        if daftar[mid]['nilai'] == target:
            return daftar[mid]
        elif daftar[mid]['nilai'] < target:
            right = mid + 1
        else:
            left = mid - 1
    return None

daftar_mahasiswa = []

print("== Input Data Mahasiswa ==")
jumlah = int(input("masukkan jumlah mahasiswa: "))

for i in range(jumlah):
    print(f"data ke-{i+1}:")
    nama = input("masukkan Nama : ")
    nilai = int(input("masukkan Nilai: "))
    daftar_mahasiswa.append({'nama': nama, 'nilai': nilai})

daftar_mahasiswa.sort(key=lambda x: x['nilai'])

print("== Daftar Mahasiswa (diurutkan berdasarkan nilai) ==")
for mhs in daftar_mahasiswa:
    print(f"Nama: {mhs['nama']:<15} | Nilai: {mhs['nilai']}")


print("== Cari Nilai ==")
cari = int(input("masukkan nilai yang ingin dicari: "))

hasil = binary_search_nilai(daftar_mahasiswa, cari)

if hasil:
    print(f"HASIL: Nilai {cari} ditemukan! Pemiliknya adalah {hasil['nama']}.")
else:
    print(f"HASIL: Nilai {cari} tidak ditemukan dalam daftar.")
```
